```python
import numpy as np
import plotly.express as px
```

```python
def function(x,upper,lower):
    return (upper-lower)*x/(x**2+(upper+lower)*x+(upper*lower))
def derivative(x,upper,lower):
    return(lower-upper)*(x**2-upper*lower)/(x**2+(upper+lower)*x+upper*lower)**2
```

```python
upper = 163000
lower = 30000

x = np.linspace(1000,500000,20000)
y = function(x,upper,lower)
y_derivative = derivative(x,upper,lower) 
```

```python
px.scatter(x=x,y=y, render_mode='webgl')
```

```python
px.scatter(x=x,y=y_derivative, render_mode='webgl')
```
